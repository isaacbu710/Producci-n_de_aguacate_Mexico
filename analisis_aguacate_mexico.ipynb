```python
# ---------------------------------------
# Anlisis de Produccin de Aguacate
# ---------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ruta absoluta del CSV
ruta_csv = r"C:\Users\herna\Producci-n_de_aguacate_Mexico\data\world food production.csv"

# Verificar que el archivo exista
if not os.path.exists(ruta_csv):
    raise FileNotFoundError(f"No se encontr el archivo: {ruta_csv}")

# Cargar dataset
data = pd.read_csv(ruta_csv)

# Revisar columnas
print("Columnas del dataset:", data.columns.tolist())

# Filtrar solo aguacate
avocado = data[["Entity", "Year", "Avocados Production (tonnes)"]]

# Seleccionar pases de inters
paises = ["Mexico", "Colombia", "Peru", "Dominican Republic", "Indonesia", "Kenya"]
avocado = avocado[avocado["Entity"].isin(paises)]

# Verificar datos
display(avocado.head())

# Configuracin de grficos
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")

# Grfico de lnea
sns.lineplot(data=avocado, x="Year", y="Avocados Production (tonnes)", hue="Entity", marker="o")
plt.title("Produccin de Aguacate: Mxico y Principales Pases Productores", fontsize=14)
plt.xlabel("Ao", fontsize=12)
plt.ylabel("Produccin (toneladas)", fontsize=12)
plt.legend(title="Pas")
plt.xticks(rotation=45)
plt.tight_layout()

# Crear carpeta de imgenes si no existe
ruta_images = r"C:\Users\herna\Producci-n_de_aguacate_Mexico\images"
if not os.path.exists(ruta_images):
    os.makedirs(ruta_images)

# Guardar grfico
plt.savefig(os.path.join(ruta_images, "grafico_produccion.png"))

plt.show()
```
